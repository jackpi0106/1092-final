```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

data1 = pd.read_csv("發電裝置容量1.csv", encoding = 'Big5')
data2 = pd.read_csv("聯合再生收盤價.csv", encoding = 'Big5')
data3 = pd.read_csv("中興電收盤價.csv", encoding = 'Big5')
data4 = pd.read_csv("世紀鋼收盤價.csv", encoding = 'Big5')
data5 = pd.read_csv("茂迪收盤價.csv", encoding = 'Big5')

plt.rcParams['font.sans-serif']=['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(17,5))
plt.plot(data2['年月'], data2['收盤'], label='聯合再生', color="red")
plt.plot(data3['年月'], data3['收盤'], label='中興電', color="green")
plt.plot(data4['年月'], data4['收盤'], label='世紀鋼', color="blue")
plt.plot(data5['年月'], data5['收盤'], label='茂迪', color="gray")
plt.legend()
plt.title("相關概念股趨勢")
plt.xlabel("年月(201603-202103)")
plt.ylabel("收盤價")
plt.xticks(rotation=45) 
plt.show()
```

```python
plt.figure(figsize=(17,5))
plt.plot(data1['年月'], data1['太陽光電'], label='太陽光電', color="red")
plt.plot(data1['年月'], data1['風力'], label='風力', color="green")
plt.legend()
plt.title("發電量增減趨勢")
plt.xlabel("年月(201603-202103)")
plt.ylabel("發電裝置容量(千瓩)")
plt.xticks(rotation=45)
plt.show()
```
