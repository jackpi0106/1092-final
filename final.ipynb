```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator

data1 = pd.read_csv("1.csv", encoding = 'Big5')
data2 = pd.read_csv(".csv", encoding = 'Big5')
data3 = pd.read_csv(".csv", encoding = 'Big5')
data4 = pd.read_csv(".csv", encoding = 'Big5')
data5 = pd.read_csv(".csv", encoding = 'Big5')

plt.rcParams['font.sans-serif']=['Microsoft JhengHei']
plt.rcParams['axes.unicode_minus']=False

plt.figure(figsize=(17,5))
plt.plot(data2[''], data2[''], label='', color="red")
plt.plot(data3[''], data3[''], label='', color="green")
plt.plot(data4[''], data4[''], label='', color="blue")
plt.plot(data5[''], data5[''], label='', color="gray")
plt.legend()
plt.title("")
plt.xlabel("(201603-202103)")
plt.ylabel("")
plt.xticks(rotation=45) 
plt.show()
```
