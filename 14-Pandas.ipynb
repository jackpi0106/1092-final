## Filtering Data

單一判斷

```python
age24 = csvdata[csvdata['Age'] > 23]
age24
```

多重判斷

```python
score90 = csvdata[(csvdata['Subject1'] >= 90)&(csvdata['Subject2'] >= 90)]
score90
```

iloc: 以數字為 index 來做選擇

```python
csvdata.iloc[0:5]
```

```python
#找出 index 3 行第0欄的值
csvdata.iloc[0,6]
```

loc: 以 column 名來做選擇

```python
#找出所有今年剛滿 18 歲的人
a=csvdata.loc[csvdata['Age'] == 18]
a
```

從 County 取出縣市

```python
num = int(len(csvdata))
i=0
for i in range(num):
    csvdata['New County'] = csvdata['County'].apply(lambda x: x[:3])
csvdata
```

依據各縣市計算該縣市Subject1的平均

```python
csvdata.groupby(by='New County')['Subject1'].mean()
```
